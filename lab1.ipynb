```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Diabets_World.csv", sep=',', encoding='utf-8')
df.head()
```

```python
my_variant = 26
arr = df.COUNTRY.unique()
data = {i+1: arr[i] for i in range(len(arr))}
print(arr)
print(data)
variant = data[my_variant]
print(f"Мій варіант: {variant}")
```

```python
filtered_data = df.loc[df.COUNTRY == variant]
filtered_data = filtered_data.reset_index(drop=True)
filtered_data
```

# 1 завдання

```python
df = pd.DataFrame(filtered_data)

years = df['YEAR']
values = df['VALUE']

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(years, values, marker='o', linestyle='-', color='b')
plt.xlabel('Рік')
plt.ylabel('Значення')
plt.title('Графік множини спостережень')
plt.grid(True)
plt.show()
```

# 2 завдання

```python
# Розділення вибірки на дослідну та тестову частини (80% для дослідної та 20% для тестової)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Побудова графіку
plt.figure(figsize=(10, 6))
plt.scatter(train_data['YEAR'], train_data['VALUE'], color='blue', label='Дослідна вибірка')
plt.scatter(test_data['YEAR'], test_data['VALUE'], color='red', label='Тестова вибірка')
plt.title('Розділена вибірка на дослідну та тестову частини')
plt.xlabel('Рік')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)
plt.show()
```

# 3 завдання

```python
# Знаходження точки центру мас для дослідної вибірки
center_of_mass = {
    'YEAR': train_data['YEAR'].mean(),
    'VALUE': train_data['VALUE'].mean()
}

# Побудова графіку з точкою центру мас
plt.figure(figsize=(10, 6))
plt.scatter(train_data['YEAR'], train_data['VALUE'], color='blue', label='Дослідна вибірка')
plt.scatter(test_data['YEAR'], test_data['VALUE'], color='red', label='Тестова вибірка')
plt.scatter(center_of_mass['YEAR'], center_of_mass['VALUE'], color='green', marker='X', s=200, label='Точка центру мас')
plt.title('Розділена вибірка на дослідну та тестову частини з точкою центру мас')
plt.xlabel('Рік')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)
plt.show()

# Вивід координат точки центру мас
print(f"Точка центру мас: Рік - {center_of_mass['YEAR']}, Значення - {center_of_mass['VALUE']}")
```

### b) Програмний код для знаходження коефіцієнтів лінійної регресії:

```python
def linear_regression_coeffs(x, y):
    N = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x[i] * y[i] for i in range(N))
    sum_x_squared = sum(x[i] ** 2 for i in range(N))
    
    m = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x ** 2)
    b = (sum_y - m * sum_x) / N
    
    return m, b

# Розрахунок коефіцієнтів лінійної регресії для дослідної вибірки
train_data = train_data.reset_index(drop=True)

m, b = linear_regression_coeffs(train_data['YEAR'], train_data['VALUE'])
print(f'Коефіцієнти лінійної регресії: m = {m}, b = {b}')
```

### d. Таблиця зі значеннями коефіцієнтів лінійних регресій і їх сумарні похибки:

```python
# Розрахунок сумарної похибки
def calculate_mse(m, b, x, y):
    N = len(x)
    return sum((y[i] - (m * x[i] + b)) ** 2 for i in range(N)) / N

# Збір даних для таблиці
m, b = linear_regression_coeffs(train_data['YEAR'], train_data['VALUE'])

results = []
for i in range(1, 11):
    m_i = i * 0.1
    b_i = b
    mse_i = calculate_mse(m_i, b_i, train_data['YEAR'], train_data['VALUE'])
    results.append({'m': m_i, 'b': b_i, 'MSE': mse_i})

# Створення DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```

### e) Побудова 10 графіків лінійних регресій:

```python
plt.figure(figsize=(10, 5))

for i, row in results_df.iterrows():
    m_i = row['m']
    b_i = row['b']
    plt.plot(train_data['YEAR'], m_i * train_data['YEAR'] + b_i, label=f'Лінійна регресія {i+1}: y = {m_i:.2f}x + {b_i:.2f}')

plt.scatter(train_data['YEAR'], train_data['VALUE'], color='blue', label='Дослідна вибірка')
plt.title('Лінійні регресії для дослідної вибірки')
plt.xlabel('Рік')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)
plt.show()
```

# 5 завдання

```python
# Знаходження індексу рядка з найменшою сумарною квадратичною похибкою
optimal_idx = results_df['MSE'].idxmin()

# Отримання оптимальних значень m та b
optimal_m = results_df.loc[optimal_idx, 'm']
optimal_b = b  # Використання b, яке ми отримали з попередніх розрахунків

# Вивід оптимальних значень
print(f'Оптимальні значення: m = {optimal_m}, b = {optimal_b}')
# Відображення дослідних данних
plt.scatter(train_data['YEAR'], train_data['VALUE'], label='Дослідні дані')

# Побудува оптимальної лінійної регресії
y_pred = optimal_m * train_data['YEAR'] + optimal_b
plt.plot(train_data['YEAR'], y_pred, color='red', label=f'Лінійна регресія (m={optimal_m:.2f}, b={optimal_b:.2f})')

plt.xlabel('Рік')
plt.ylabel('Значення')
plt.legend()
plt.show()
```

# 6 завдання

```python
# Застосуйте оптимальні значення m та b до тестової вибірки
test_data = test_data.reset_index(drop=True)

test_predictions = optimal_m * test_data['YEAR'] + optimal_b

# Розрахунок сумарної похибки
test_mse = calculate_mse(optimal_m, optimal_b, test_data['YEAR'], test_predictions)

# Створення таблиці для представлення даних
test_results = pd.DataFrame({'m': [optimal_m], 'b': [optimal_b], 'MSE': [test_mse]})
test_results
```
